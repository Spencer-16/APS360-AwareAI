```python
import os
import numpy as np
from PIL import Image

# Iterate over each file in the directory
def calculate_max_depth_and_rename_file(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.png'):
            file_path = os.path.join(directory, filename)

            # Open the image
            image = Image.open(file_path)

            # Convert the image to grayscale
            image = image.convert('L')

            # Convert image to numpy array
            img_array = np.array(image)

            # Calculate the bottom 25% of the image along the y-axis
            y_start = int(0.75 * img_array.shape[0])
            bottom_region = img_array[y_start:, :]

            # Find the maximum pixel value in the bottom region
            max_pixel = np.max(bottom_region)

            # Rename the file to the maximum depth value
            #new_filename = f'{max_pixel}.png'
            #new_file_path = os.path.join(directory, new_filename)

            # Rename the file
            #os.rename(file_path, new_file_path)

            print(f'original "{filename}" depth value:"{max_pixel}"')

```

```python
# Specify the folder containing the NPY files
directory = '/content/gdrive/MyDrive/3rd_yr/AI/APS360/SVM/dataset_depth'
calculate_max_depth_and_rename_file(directory)

```
