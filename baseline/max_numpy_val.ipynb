```python
import os
import numpy as np
from PIL import Image
import shutil
from google.colab import drive
drive.mount('/content/gdrive')

def calculate_max_pixel_value(file_path):
    try:
        # Load the NumPy array from the file
        array = np.load(file_path)

        # Calculate the maximum pixel value
        max_pixel_value = np.max(array)

        return max_pixel_value
    except IOError:
        print("Error: Unable to load the file.")
    except ValueError:
        print("Error: Invalid data in the NumPy array.")
    except Exception as e:
        print("An error occurred:", str(e))


# Example usage
file_path = '/content/gdrive/MyDrive/3rd_yr/AI/APS360/SVM/test_input/test.npy'
max_value = calculate_max_pixel_value(file_path)
print("Maximum pixel value:", max_value)
```
