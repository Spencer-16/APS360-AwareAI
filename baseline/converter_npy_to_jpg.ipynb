```python
import os
import numpy as np
from PIL import Image
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
def convert_npy_to_jpg(input_folder, output_folder):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    # Get a list of all files in the input folder
    files = os.listdir(input_folder)

    # Loop through each file in the input folder
    for file_name in files:
        # Construct the full path to the input file
        input_path = os.path.join(input_folder, file_name)

        # Check if the file is an NPY file
        if os.path.isfile(input_path) and file_name.lower().endswith('.npy'):
            # Load the NPY file
            data = np.load(input_path)

            # Normalize data (if needed)
            normalized_data = (data - np.min(data)) / (np.max(data) - np.min(data)) * 255

            # Convert numerical array to image
            image = Image.fromarray(normalized_data.astype(np.uint8))

            # Construct the full path to the output file
            output_path = os.path.join(output_folder, os.path.splitext(file_name)[0] + '.jpg')

            # Save the image as JPEG
            image.save(output_path)
            print(f"Converted {file_name} to JPEG.")
```

```python
# Specify the input and output folders
input_dir = '/content/gdrive/MyDrive/3rd_yr/AI/APS360/SVM/test_input'
output_dir = '/content/gdrive/MyDrive/3rd_yr/AI/APS360/SVM/test_output'

# Convert images to JPEG
convert_npy_to_jpg(input_dir, output_dir)
```
